### Joining Categories into 1 list

```python
ex = df['category_list'].iloc[0]
print(ex)
ex.strip('|')
```

```python
## Strip l/r | befre splitting and re-joining with ,
res = df['category_list'].astype('str').apply(lambda x: ','.join(x.strip('|').split('|')))
res = list(res.values)
res[:10]
```

```python
## Join lists and combin into one comma-sep list
text = ','.join(res)
## split into list then take unique values as list
category_list = list(set(text.split(',')))
category_list[:10]
```
